# Instructions.
    # On my personal twitter page, the user can tweet a phrase to get a graphed sentiment analysis of the
      specified twitter handle.  The input must be in the format: "New Tweet Analysis: @username".
    # The program will perform five cycles, one cycle every five minutes.
    # If the user enters a request while the program is running, the request will be handled in the next 
      cycle of the program.
    # Only unique plots are grabbed from the mentions section of my twitter page (direct tweets).
    # Plots will be added according to chronological order of my timeline.
    # The user's name will be mentioned in the return tweet text.

```python
# Dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import json
import tweepy
import time
from pprint import pprint

# Initialize Sentiment Analyzer
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()

# Twitter API Keys
consumer_key = ""
consumer_secret = ""
access_token = ""
access_token_secret = ""
```

### Define the function.

```python
def plot_bot(target_user):
    auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
    auth.set_access_token(access_token, access_token_secret)
    api = tweepy.API(auth)
    
    oldest_tweet = ""
    counter = 1
    sentiments = []
    
    for page in tweepy.Cursor(
        api.search,
        q = target_user, 
        count = 100,
        max_id = oldest_tweet,
        result_type = "recent").pages(5):
    
        for tweet in page:
            compound = analyzer.polarity_scores(tweet.text)["compound"]
            positive = analyzer.polarity_scores(tweet.text)["pos"]
            neutral = analyzer.polarity_scores(tweet.text)["neu"]
            negative = analyzer.polarity_scores(tweet.text)["neg"]
        
            sentiments.append({
                "Date": tweet.created_at,
                "Compound": compound,
                "Positive": positive,
                "Neutral": neutral,
                "Negative": negative,
                "Tweets Ago": counter
            })
        
            counter += 1
    df = pd.DataFrame(sentiments)
    
    plt.figure(figsize = (10, 7))
    plt.plot(-df["Tweets Ago"], df["Compound"], marker = "o", color = "navy", linewidth = 0.5)
    plt.legend({target_user}, loc = "center left", bbox_to_anchor=(1.02,0.5), ncol=1, shadow=True)

    plt.title("Sentiment Analysis of Tweets")
    plt.xlabel("Tweets Ago")
    plt.ylabel("Tweet Polarity")
    plt.xlim(-500, 0)
    plt.ylim(-1,1)
    plt.grid()
    plt.savefig("Figure1.jpg", bbox_inches = "tight")
    plt.show()
    
    api.update_with_media("Figure1.jpg", f"Thanks {user}!")
```

### Run the program, graph the output.

```python
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth, parser = tweepy.parsers.JSONParser())


test = []
test1 = []
texts = []

counter1 = 0
while counter1 < 5:
    public_tweets = api.mentions_timeline()
    for tweet in reversed(public_tweets):
        if "New Tweet Analysis:" in tweet["text"]:

            if (tweet["text"][34:]) not in test:
                plot_bot(tweet["text"][34:])
                test.append(tweet["text"][34:])
                user = tweet["user"]["screen_name"]
                test.sort()
                test1.sort()
    time.sleep(300)
    counter1 += 1
    print ("Five minutes have elapsed.")
        
```
