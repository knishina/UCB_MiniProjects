```python
import pandas as pd
import numpy as np
import requests
import json
from pprint import pprint
from config import api_key
import matplotlib.pyplot as plt
import random
import time
time.sleep(1)
```

```python
# Import the csv.  Look for unique values via longitude/latitude.
# Export it as working_file.csv.
datafile1 = "worldcities.csv"
df_unique = pd.read_csv(datafile1)

df = df_unique.drop_duplicates(subset = ["Latitude", "Longitude"], keep = False)
df.to_csv("working_file.csv")
datafile = "working_file.csv"
```

```python
# Get 500 randomly picked entries.
n = sum(1 for line in open(datafile))-1
s = 500
skip = sorted(random.sample(range(1, n+1), n-s))
df = pd.read_csv(datafile, skiprows=skip)
df.head()
```

```python
# Data grab!
df["Temperature (F)"] = ""
df["Humidity (%)"] = ""
df["Cloudiness (%)"] = ""
df["Wind Speed (mph)"] = ""

for index, row in df.iterrows():
    url = "https://api.openweathermap.org/data/2.5/weather?q="
    end = "&units=imperial&APPID=" + api_key

    city = row["City"]
    country = row["Country"]
    location = city + ", " + country
    
    final_url = url + location + end
    print (final_url)
    
    response = requests.get(final_url).json()
    try:
        df.set_value(index, "Temperature (F)", response["main"]["temp_max"])
        df.set_value(index, "Humidity (%)", response["main"]["humidity"])
        df.set_value(index, "Cloudiness (%)", response["clouds"]["all"])
        df.set_value(index, "Wind Speed (mph)", response["wind"]["speed"])
    except:
        print ("0")
```

```python
# See what I've got.  Blank spaces...
df.to_csv("Koren.csv")
df.head(10)
```

```python
# Fill in the blank spaces with NaN so that they aren't graphed.
df["Temperature (F)"] = df["Temperature (F)"].replace(r"", np.NaN)
df["Humidity (%)"] = df["Humidity (%)"].replace(r"", np.NaN)
df["Cloudiness (%)"] = df["Cloudiness (%)"].replace(r"", np.NaN)
df["Wind Speed (mph)"] = df["Wind Speed (mph)"].replace(r"", np.NaN)
df.head(10)
```

```python
plt.figure(figsize = (10, 7))
plt.xlim(-80, 100)
plt.ylim(-100, 150)
plt.scatter(df["Latitude"], df["Temperature (F)"], marker = "o", facecolors = "blue", edgecolors = "black")
plt.title("City Latitude vs. Max Temperature")
plt.ylabel("Max Temperature (F)")
plt.xlabel("Latitude")
plt.grid()
plt.show()
```

```python
plt.figure(figsize = (10, 7))
plt.xlim(-80, 100)
plt.ylim(-20, 120)
plt.scatter(df["Latitude"], df["Humidity (%)"], marker = "o", facecolors = "blue", edgecolors = "black")
plt.title("City Latitude vs. Humidity")
plt.xlabel("Latitude")
plt.ylabel("Humidity (%)")
plt.grid()
plt.show()
```

```python
plt.figure(figsize = (10, 7))
plt.xlim(-80, 100)
plt.ylim(-20, 120)
plt.scatter(df["Latitude"], df["Cloudiness (%)"], marker = "o", facecolors = "blue", edgecolors = "black")
plt.title("City Latitude vs. Cloudiness")
plt.xlabel("Latitude")
plt.ylabel("Cloudiness (%)")
plt.grid()
plt.show()
```

```python
plt.figure(figsize = (10, 7))
plt.xlim(-80, 100)
plt.ylim(-5, 40)
plt.scatter(df["Latitude"], df["Wind Speed (mph)"], marker = "o", facecolors = "blue", edgecolors = "black")
plt.title("City Latitude vs. Wind Speed")
plt.xlabel("Latitude")
plt.ylabel("Wind Speed (mph)")
plt.grid()
plt.show()
```
