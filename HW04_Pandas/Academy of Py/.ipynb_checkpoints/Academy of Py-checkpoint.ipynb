```python
import pandas as pd
datafile1 = "raw_data/schools_complete.csv"
datafile2 = "raw_data/students_complete.csv"
schools = pd.read_csv(datafile1)
students = pd.read_csv(datafile2)
```

```python
# Create a high level snapshot (in table form) of the district's key metrics, including:
    # Total Schools
    # Total Students
    # Total Budget
    # Average Math Score
    # Average Reading Score
    # % Passing Math
    # % Passing Reading
    # Overall Passing Rate (Average of Passing Math and Passing Reading.)
```

```python
tot_schools = len(schools)
tot_students = schools["size"].sum()
tot_budget = schools["budget"].sum()
ave_math_score = round((students["math_score"].mean()), 2)
ave_reading_score = round((students["reading_score"].mean()), 2)

math_above70 = students.loc[students["math_score"] >= 70].count()
math_percent = round(((math_above70["math_score"]/tot_students) * 100), 2)

reading_above70 = students.loc[students["reading_score"] >= 70].count()
reading_percent = round((reading_above70["reading_score"]/tot_students * 100), 2)

overall_average = round(((ave_math_score + ave_reading_score)/2), 2)
overall_average

high_level = pd.DataFrame({
    "Total Schools": [tot_schools],
    "Total Students": [tot_students],
    "Total Budget": [tot_budget],
    "Average Math Score": [ave_math_score],
    "Average Reading Score": [ave_reading_score],
    "% Passing Math": [math_percent],
    "% Passing Reading": [reading_percent],
    "% Overall Passing Rate": [overall_average]
})

high_level["Total Students"] = high_level["Total Students"].map("{:,}".format)
high_level["Total Budget"] = high_level["Total Budget"].map("${:,.2f}".format)

high_level[[
    "Total Schools",
    "Total Students",
    "Total Budget",
    "Average Math Score",
    "Average Reading Score",
    "% Passing Math",
    "% Passing Reading",
    "% Overall Passing Rate"
]]
```

```python
# Create an overview table that summarizes key metrics about each school, including:
    # School Name
    # School Type
    # Total Students
    # Total School Budget
    # Per Student Budget
    # Average Math Score
    # Average Reading Score
    # % Passing Math
    # % Passing Reading
    # Overall Passing Rate (average of the above two)
```

```python
# Table #1
schools["Per Student Budget"] = round((schools["budget"]/schools["size"]), 2)
schools["Per Student Budget"] = schools["Per Student Budget"].map("${:,.2f}".format)
schools["budget"] = schools["budget"].map("${:,.2f}".format)
schools["Total Students"] = schools["size"].map("{:,}".format)


schools_r = schools.rename(columns = {
    "name": "School Name",
    "budget": "Total School Budget",
    "type": "School Type"
})
schools_re = schools_r.sort_values("School Name", ascending = True)
schools_ren = schools_re.reset_index()
schools_rena = schools_ren.drop(["index", "School ID", "size"], axis = 1)
schools_rename = pd.DataFrame(schools_rena)


# Table #2
school_sort = students.groupby(["school"])
list_of_schools = sorted(list(students["school"].unique()))

tot_students = list(school_sort["school"].value_counts())
ave_mscore = list(round((school_sort["math_score"].mean()), 2))
ave_rscore = list(round((school_sort["reading_score"].mean()), 2))

students_new_m = students.loc[students["math_score"] >= 70]
snm_sort = students_new_m.groupby(["school"])
snm_count = snm_sort["math_score"].count()
passing_math = list(round(((snm_count/tot_students) * 100), 2))

students_new_r = students.loc[students["reading_score"] >= 70]
snr_sort = students_new_r.groupby(["school"])
snr_count = snr_sort["reading_score"].count()
passing_reading = list(round(((snr_count/tot_students) * 100), 2))

columns = ["School Name", "Average Math Score", "Average Reading Score", "% Passing Math", "% Passing Reading", "% Overall Passing"]
df_new = pd.DataFrame(columns = columns)
df_new["School Name"] = list_of_schools
df_new["Average Math Score"] = ave_mscore
df_new["Average Reading Score"] = ave_rscore
df_new["% Passing Math"] = passing_math
df_new["% Passing Reading"] = passing_reading

overall_pass = round(((df_new["% Passing Math"] + df_new["% Passing Reading"])/2), 2)
df_new["% Overall Passing"] = overall_pass

df_new = df_new.sort_values("School Name", ascending = True)
df_new = pd.DataFrame(df_new)

# Merge the two tables.
df_summary = pd.merge(schools_rename, df_new, on = "School Name", right_index = True)
df_summary = pd.DataFrame(df_summary)
df_summary.set_index("School Name")
```

```python
# Create a table that lists the average math score for students of each grade
# level at each school.
```

```python
math_group = students.groupby(["school", "grade"])
math_averages = math_group["math_score"].mean()

math_by_grade_unstacked = math_averages.unstack(level = 1)
columns1 = ["9th", "10th", "11th", "12th"]

math_g = math_by_grade_unstacked[columns1]
math_g = math_g.style.format({
    "9th": "{:.2f}".format,
    "10th": "{:.2f}".format, 
    "11th": "{:.2f}".format, 
    "12th": "{:.2f}".format
})


print ("Math scores by school for each grade level.")
math_g
```

```python
reading_averages = math_group["reading_score"].mean()

reading_by_grade_unstacked = reading_averages.unstack(level = 1)
columns1 = ["9th", "10th", "11th", "12th"]

reading_g = reading_by_grade_unstacked[columns1]
reading_g = reading_g.style.format({
    "9th": "{:.2f}".format,
    "10th": "{:.2f}".format, 
    "11th": "{:.2f}".format, 
    "12th": "{:.2f}".format
})


print ("Reading scores by school for each grade level.")
reading_g
```

```python
# Create a table that breaks down school performances based on average 
# Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. 
# Include in the table each of the following:
    # Average Math Score
    # Average Reading Score
    # % Passing Math
    # % Passing Reading
```

```python
schools_rename["Per Student Budget"] = schools_rename["Per Student Budget"].map(lambda k:k.replace("$", "")).astype(float)

bins = [570, 592.5, 615, 637.5, 660]
group_names = ["1 (Improvement Needed)", "2 (Meets Expectations)", "3 (Exceeds Expectations)", "4 (Exceptional)"]
group = pd.cut(schools_rename["Per Student Budget"], bins, labels = group_names)
schools_rename["Spending Ranges (per Student)"] = group

rating_df = schools_rename[["School Name", "Spending Ranges (per Student)"]]

merged_rat = pd.merge(rating_df, df_new, on = "School Name")
merged_rat = merged_rat.sort_values(["Spending Ranges (per Student)"], ascending = True)
del merged_rat["School Name"]

group1 = merged_rat.groupby(["Spending Ranges (per Student)"])
group2 = group1["Average Math Score", "Average Reading Score", "% Passing Math", "% Passing Reading", "% Overall Passing"].mean()
group2.style.format({
    "Average Math Score": "{:.2f}".format,
    "Average Reading Score": "{:.2f}".format,
    "% Passing Math": "{:.2f}".format,
    "% Passing Reading": "{:.2f}".format,
    "% Overall Passing": "{:.2f}".format
})
```

```python
# Repeat the above breakdown, but this time group schools based on a reasonable 
# approximation of school size (Small, Medium, Large).
```

```python
schools_rename["Total Students"] = schools_rename["Total Students"].map(lambda k:k.replace(",", "")).astype(int)


bins1 = [425, 1950, 3475, 5000]
group_names1 = ["Small", "Medium", "Large"]
groupa = pd.cut(schools_rename["Total Students"], bins1, labels = group_names1)
schools_rename["School Size"] = groupa

rating_df = schools_rename[["School Name", "School Size"]]

merged_rat = pd.merge(rating_df, df_new, on = "School Name")
merged_rat = merged_rat.sort_values(["School Size"], ascending = True)
del merged_rat["School Name"]

groupb = merged_rat.groupby(["School Size"])
groupc = groupb["Average Math Score", "Average Reading Score", "% Passing Math", "% Passing Reading", "% Overall Passing"].mean()
groupc.style.format({
    "Average Math Score": "{:.2f}".format,
    "Average Reading Score": "{:.2f}".format,
    "% Passing Math": "{:.2f}".format,
    "% Passing Reading": "{:.2f}".format,
    "% Overall Passing": "{:.2f}".format
})
```

```python
# Repeat the above breakdown, but this time group schools based on school type 
# (Charter vs. District).
```

```python
rating_df = schools_rename[["School Name", "School Type"]]

merged_rat = pd.merge(rating_df, df_new, on = "School Name")
merged_rat = merged_rat.sort_values(["School Type"], ascending = True)
del merged_rat["School Name"]

group1a = merged_rat.groupby(["School Type"])
group2a = group1a["Average Math Score", "Average Reading Score", "% Passing Math", "% Passing Reading", "% Overall Passing"].mean()
group2a.style.format({
    "Average Math Score": "{:.2f}".format,
    "Average Reading Score": "{:.2f}".format,
    "% Passing Math": "{:.2f}".format,
    "% Passing Reading": "{:.2f}".format,
    "% Overall Passing": "{:.2f}".format
})
```

```python
# Three Observable Trends:
    # The first trend is that spending per student is inversely proportional to the overall
    # passing percentage.  This means that the more that is spent per student, the lower the
    # overall passing rate of the students.
    
    # The second trend is that the smaller the school, the better the overall passing
    # percentage.
    
    # The third trend is that charter schools perform significantly better than district 
    # schools.  In fact, the top five performers are charter schools; the bottom five are 
    # district schools.
```
